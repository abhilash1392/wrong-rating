```python
import pandas as pd 
import numpy as np 
import string 
from nltk.corpus import stopwords 
from nltk.tokenize import WhitespaceTokenizer,word_tokenize
from nltk.stem import WordNetLemmatizer 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('chrome_reviews.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df = df[['Text','Star']]
df.head()
```

```python
df = df.dropna(axis=0).reset_index(drop=True)
```

```python
df['is_negative_review'] = df['Star'].apply(lambda x:1 if x<3 else 0)
df.head()
```

```python
## Cleaning the text 
stopword = stopwords.words('english')
new_words = ['google','chrome','app','browser','apps','update','updated','aap']
stopword.extend(new_words)
import re 

import emoji

def give_emoji_free_text(text):
    return emoji.get_emoji_regexp().sub(r'', text)

def clean_text(text):
    text = give_emoji_free_text(text)
    text = "".join([word.lower() for word in text if word not in string.punctuation 
            and not word.isdigit()])
    text = word_tokenize(text)
    text = [word for word in text if word not in new_words]
    text = [WordNetLemmatizer().lemmatize(word) for word in text]
    text = " ".join(text)
    return text
```

```python
print(clean_text('This is very helpfull aap'))
```

```python
print(clean_text('Not able to update'))
```

```python
df["TextClean"] = df["Text"].apply(lambda x: clean_text(x))
```

```python
df.head()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()
df['sentiments'] = df['TextClean'].apply(lambda x: sid.polarity_scores(x))
df  = pd.concat([df.drop(['sentiments'],axis=1),df['sentiments'].apply(pd.Series)],axis=1)
df.head()
```

```python
df['nb_chars'] = df['TextClean'].apply(lambda x: len(x))
df['nb_words'] = df['TextClean'].apply(lambda x: len(x.split(" ")))
df.head()
```

```python
from gensim.test.utils import common_texts 
from gensim.models.doc2vec import Doc2Vec, TaggedDocument 

documents = [TaggedDocument(doc,[i]) for i,doc 
             in enumerate(df['TextClean'].apply(lambda x : x.split(" ")))]

model = Doc2Vec(documents, vector_size = 5 , window=2, min_count = 1 ,workers = 4)
doc2vec_df = df["TextClean"].apply(lambda x:model.infer_vector(x.split(" "))).apply(pd.Series)
doc2vec_df.columns = ["doc2vec_vector_"+str(x) for x in doc2vec_df.columns]
df = pd.concat([df,doc2vec_df],axis=1)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
tfidf = TfidfVectorizer(min_df = 10)
tfidf_result = tfidf.fit_transform(df["TextClean"]).toarray()
tfidf_df = pd.DataFrame(tfidf_result, columns = tfidf.get_feature_names())
tfidf_df.columns = ["word_" + str(x) for x in tfidf_df.columns]
tfidf_df.index = df.index 
df = pd.concat([df,tfidf_df],axis=1)
df.head()
```

```python
df["is_negative_review"].value_counts(normalize=True)
```

```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
    background_color = 'white',
    max_words = 200,
    max_font_size= 40,
    scale=3,
    random_state=42).generate(str(data))
    fig = plt.figure(1, figsize =(20,20))
    plt.axis('off')
    if title:
        fig.suptitle(title,fontsize=20)
        fig.subplots_adjust(top = 2.3)
        
    plt.imshow(wordcloud)
    plt.show()
    
show_wordcloud(df["TextClean"])
```

```python
 
import seaborn as sns
 
for x in [0, 1]:
    subset = df[df['is_negative_review'] == x]
    
    # Draw the density plot
    if x == 0:
        label = "Good reviews"
    else:
        label = "Bad reviews"
    sns.distplot(subset['compound'], hist = False, label = label)
```

```python
label = 'is_negative_review'
ignore_cols = ["Star",label,"Text","TextClean","nb_words","nb_chars"]
features = [c for c in df.columns if c not in ignore_cols]

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
y = kmeans.fit_predict(df[features])
y
```

```python
df["Predicted"] = y
df["Predicted"].value_counts()
```
